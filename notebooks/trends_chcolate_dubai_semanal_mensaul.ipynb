Con este bloque de código, definimos distintas variables para chocolate Dubái, pistacho y masa Kataifi. Agrupamos la fuerza de tendencia en tres componentes que forman parte de la tableta de chocolate Dubái. El objetivo es realizar una comparativa de tendencia y fechas respecto a nuestro trabajo previo con las cinco marcas analizadas en el archivo .csv anterior. Se van a generar dos archivos distintos tipo .csv, uno para fuerzas de tendencias semanales y otro mensuales.

```python
# 1 Se instalan y se cargan las librerías a usar.
!pip install "urllib3<2" pytrends
import pandas as pd
from pytrends.request import TrendReq
import time

pytrends = TrendReq(hl='es-ES', tz=360, retries=10, backoff_factor=2)

# 2 Se crean las variables y se ejecuta la búsqueda.
conceptos = {
   "trend_chocolate_dubai": "Chocolate Dubai + Chocolate de Dubai + Dubai Chocolate + Tableta Dubai",
   "trend_pistacho": "Pistacho + Crema de pistacho + Crema pistacho + Pasta de pistacho",
   "trend_kataifi": "Kataifi + Katayfi + Kadayif + Kadaif + Fideos turcos"
}


all_dfs = []


for nombre_columna, query in conceptos.items():
   try:
       # Se solicitan los datos a Google Trends.
       pytrends.build_payload(
           [query],
           cat=0,
           timeframe='2024-01-01 2026-03-23',
           geo='ES'
       )
       df = pytrends.interest_over_time()


       if not df.empty:
           df = df.drop(columns=['isPartial'], errors='ignore')
           df.columns = [nombre_columna]
           all_dfs.append(df)


       time.sleep(7)


   except Exception as e:
       print(f"❌ Error en {nombre_columna}: {e}")


# 3 Se unen los datos y se generan versiones semanal y mensual
if all_dfs:
  df_semanal = pd.concat(all_dfs, axis=1).reset_index()
  df_semanal['date'] = pd.to_datetime(df_semanal['date'])

  df_semanal.to_csv('trends_chocolate_dubai_semanal.csv', index=False)

  df_mensual = df_semanal.set_index('date').resample('MS').mean().reset_index()
  df_mensual = df_mensual[df_mensual['date'] >= '2024-01-01']
  df_mensual.to_csv('trends_chocolate_dubai_mensual.csv', index=False)

  display(df_semanal.head())
  display(df_mensual.head())
else:
   print("No se pudo recolectar información.")
```
