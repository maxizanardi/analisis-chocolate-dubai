Aquí lo que hacemos es limpiar y modificar el archivo de precios históricos. Se realizan una serie de modificaciones y agregaciones para que podamos unirlo a los demás archivos y estén normalizados.

```python
#1 Importación de la librería Pandas para el manejo de datos.

import pandas as pd

#2 Cargamos el archivo.
nombre_archivo = 'Dataset_Precios_Historicos.csv'
df = pd.read_csv(nombre_archivo, sep=',')


#3 Función para normalizar precios: corregimos formato decimal y convertimos toneladas a kilos.
def transformar_datos(row):
    #Quitamos coma por punto en el precio.
    precio = float(str(row['Precio_EUR']).replace(',', '.'))
    ingrediente = str(row['Ingrediente']).lower()

    # Limpiamos la sección del cacao pasando de Tonelada a kg.
    if 'cacao' in ingrediente:
        # Pasa de Tonelada a Kilo
        return precio / 1000

    # El pistacho está correcto, así que no sufre modificaciones.
    else:
        return precio

#4 Transformación de datos: normalización de precios y formateo de fecha
df['precio_eur_kg'] = df.apply(transformar_datos, axis=1)
df['precio_eur_kg'] = df['precio_eur_kg'].round(2)
df['fecha'] = pd.to_datetime(df['Fecha'], dayfirst=True).dt.strftime('%Y-%m-01')

#5 Creamos el archivo final limpio y eliminamos columnas innecesarias
df_final = df[['fecha', 'Ingrediente', 'precio_eur_kg']]
df_final.columns = ['fecha', 'ingrediente', 'precio_eur_kg']

#6 Lo guardamos.
df_final.to_csv('Dataset_Precios_Historicos_Limpio.csv', index=False)
print("¡El archivo 'Dataset_Precios_Historicos_Limpio.csv' ya está listo!.")
```
